Algumas vezes será fundamental substituir valores NaN pela média (estatística). Outras vezes isso pode ser perigoso. A depender da situação, da suas escolhas como cientista de dados, você poderá tratá-los substituindo pela média, mediana, etc, ou simplesmente excluí-los. 

```python
df4.isnull() #isnull serve para mostrar os valores nulos. 
```

```python
df4.isnull().sum() #Como na "vida real" teremos alguns milhões de valores -  sum = somar 
```

Fillna: leia comom um "filtrar Not A Number". Para filtrar os valores válidos das colunas

```python
df4.fillna
```

```python
df5 = df4.fillna(df4.mean())
df5
```

Fillna - substituindo todos os valores NaN do nosso Data Frame pela média dos valores válidos das colunas

```python
df4.dropna(inplace = True)
df4
```
