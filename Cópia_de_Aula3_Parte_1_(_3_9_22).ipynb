Matplotlib - A biblioteca mais popular do Python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empreas = ['Apple', 'Google', 'Amazon',
           'Microsoft', 'Tencent', 
           'McDonald´s', 'Visa', 'Facebook', 'Alibaba', 
           'Louis Vuitton']
```

```python
bilhoes = [947.062, 819.573, 705.646,
           611.460, 214.023, 196.526, 
           191.032, 186.421, 169.966,
           124.273]
```

```python
plt.plot(empresas, bilhoes) #já se pode fazer um gráfico
```

```python
plt.plot(empresas, bilhoes)
plt.title ('Empreas mais ricas do mundo')
```

```python
plt.plot(empresas, bilhoes)
plt.title ('Empreas mais ricas do mundo')
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
```

```python
#Redimensionamento
plt.rcParams["figure.figsize"] = (15, 8)

plt.plot(empresas, bilhoes)
plt.xlabel('Empresas')
plt.ylabel('Bilhões')
plt.title ('Empreas mais ricas do mundo')
plt.show()
```

```python
#Conhecendo o fontsize - tamanho da fonte
plt.plot(empresas, bilhoes)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.show()
```

```python
#mudando a cor e legendando
plt.plot(empresas, bilhoes, color = 'green')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de doláres'], loc = 'right', fontsize=15,)
plt.show()
```

```python
#Aumentar a espessura da linha
plt.plot(empresas, bilhoes, color = 'green', lw=10)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de doláres'], loc = 'right', fontsize=15)
plt.show()
```

```python
#Adicionar marcador na linha
plt.plot(empresas, bilhoes, color = 'green', marker = 'o')
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de doláres'], loc = 'right', fontsize=15)
plt.show()
```

```python
#Adicionando linhas de grade
plt.plot(empresas, bilhoes, color = 'green', marker=10)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de doláres'], loc = 'right', fontsize=15)
plt.grid(True)
plt.show()
```

```python
#Mudando o tipo do gráfico e colocando a legenda acima
plt.bar(empresas, bilhoes, color = 'green') #mudou para barra (bar)
plt.xlabel('Empresas', fontsize=20)
plt.ylabel('Bilhões', fontsize=20)
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.legend(['Bilhões de doláres'], loc = 'high right', fontsize=15,)
plt.grid(True)
plt.show()
```

```python
#Gráfico de barra mais simples
plt.bar(empresas, bilhoes)
```

```python
cores = ['red', 'yellow', 'green', 'blue', 'purple', 'orange', 'magenta', 'white', 'brown', 'black']
```

```python
plt.bar(empresas, bilhoes, color = cores, edgecolor='black') 
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilhões de doláres', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
plt.style.available #opções de estilos
```

```python
plt.style.use('seaborn-dark')
plt.bar(empresas, bilhoes, color = cores, edgecolor='black') 
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilhões de doláres', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
#Barra Horizontal (barh)
```

```python
plt.style.use('seaborn-dark')
plt.barh(empresas, bilhoes, color = cores, edgecolor='black') 
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilhões de doláres', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```

```python
#Gráfico de dispersão
plt.style.use('seaborn-dark')
plt.scatter(empresas, bilhoes, color = cores, edgecolor='black') 
plt.title ('Empreas mais ricas do mundo', fontsize=25)
plt.ylabel('Bilhões de doláres', fontsize=20)
plt.xlabel('Empresas', fontsize=20)
plt.grid(True)
```
